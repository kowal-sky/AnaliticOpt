````python
def q_b_func(tank: Tank, r_q = 0.7, a = 0.5):
    return (

        4 * (tank.V_T / tank.A_i(a)) * tank.U_L * (tank.T_air - tank.T_L) / 
        (tank.d_o(a) * (1/r_q - 1))

    )

````
